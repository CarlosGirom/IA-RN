```python
#Cargar las librerias
import numpy as np
import sklearn.linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Empezamos a generar los datos
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100,1) * 2
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
```

```python
data = pd.DataFrame({'Horas de Estudio': horas_estudio.flatten(),
                     'Calificaciones': calificaciones.flatten()})

print(data)
```

```python
plt.scatter(horas_estudio, calificaciones)

plt.xlabel('Horas de Estudio')
plt.ylabel('Calificaciones')
plt.title('Relación entre Horas de Estudio y Calificaciones')

plt.show()
```

```python
# Crear y entrenar el modelo de regresión lineal
model = sklearn.linear_model.LinearRegression()
```

```python
model.fit(horas_estudio, calificaciones)
```

```python
#Realizar las predicciones del modelo entrenado
tiempo_estudio = [[13.5]]

calificacion_predicha = model.predict(tiempo_estudio)

print("La calificación predicha para {} horas de estudio es: {:.2f}".format(tiempo_estudio[0][0], calificacion_predicha[0][0]))
```
